```python
#Creating two lists, height, and weight
height = [1.73, 1.68, 1.71, 1.89, 1.79]
weight = [65.4, 59.2, 63.6, 88.4, 68.7]
```

```python
#BMI Calculation Using List
bmi_list = []
for i,j in zip(height,weight):
  bmi = j/i ** 2
  bmi_list.append(bmi)
print(bmi_list)
```

```python
#BMI Calculation Using Numpy Array
import numpy as np

arr_height=np.array(height)
arr_weight=np.array(weight)
```
